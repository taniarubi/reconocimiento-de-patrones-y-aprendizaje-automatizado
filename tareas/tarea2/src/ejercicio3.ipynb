# Experimento sobre el desbalance de datos en un dataset

```python
# Tratamiento de datos
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import make_classification

# Gráficos
import matplotlib.pyplot as plt
```

Utilizaremos **make_classification** para crear conjuntos de datos.

## Experimento 1

```python
X, y = make_classification(n_classes=2, class_sep=1.5, weights=[0.9, 0.1],
                           n_informative=3, n_redundant=1, flip_y=0,
                           n_features=20, n_clusters_per_class=1,
                           n_samples=1000, random_state=23
                          )
```

```python
# Número de observaciones por clase
unique, counts = np.unique(y, return_counts=True)
dict(zip(unique, counts))
```

```python
df = pd.DataFrame(X)
df['target'] = y
df.target.value_counts().plot(kind='bar', title='Count (target)');
```

```python
plt.scatter(X[:, 0], X[:, 1], marker='o', c=y,
            s=25, edgecolor='k')
```

## Conclusiones

Al estar jugando con varios de los parámetros notamos que aquellos que verdaderamente influyen para que un conjunto de datos sea desbalanceado son **flip_y** y **weights** mientras que el resto de los parámetros sólo parecen influir en las características de los datos (más no en la distribución de valores en cada clase).

El parámetro **weight** sirve para poder asignar la proporción de valores que tendrá cada una de las clases. De esta forma, si queremos un conjunto de datos desbalanceado, debemos ajustar estos valores para que así sea.  Por ejemplo, si tenemos $100$ datos y queremos una proporción $60-40$ entonces debemos escribir weight[0.6,0.4]

La suma de las proporciones no debe ser necesariamente $1$., y en algunos casos, cuando este valor es mayor a $1$, se regresan más valores de lo que deberían (y en un experimento pudimos ver que todas las clases toman casi los mismos valores).

En particular, encontramos que si **flip_y** es mayor a $0$ entonces  lo que hace es ir aumentando los valores de las clases (como si se estuviera balanceando) poco a poco.  Con un valor de **flip_y < 0.5** realmente el conjunto sigue desbalanceado, pero conforme el valor se acerca a $1$ entonces los valores de las clases de van balanceando.

Creo además que el número de clusters también puede influir, pero realmente sólo desvía los valores de las clases muy pero muy poquito, entonces no afecta mucho en el umbral que buscamos.
